## Модель прогнозирования оттока клиентов

### Разбить данные на обучающую и валидационную выборку функцией train_test_split()

```python
X = data.drop('Churn', axis =1)
y = data['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Обучающая - 80%, валидационная - 20%.

### Обучить модель на train-выборке логистической регрессией и случайным лесом.

```python
# Создаем функцию для расчета метрик  accuracy, precision и recall на выходе
# На входе модель, признаки и целевые переменные обучающей и валидационной выборки
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    
    print('Accuracy: {:.2f}'.format(accuracy_score(y_test, predictions)))
    print('Precision: {:.2f}'.format(precision_score(y_test, predictions)))
    print('Recall: {:.2f}'.format(recall_score(y_test, predictions)))
```

Логистическая регрессия:

```python
make_prediction(LogisticRegression(solver='liblinear',random_state=0), X_train, y_train, X_test, y_test)
```

Случайный лес:

```python
make_prediction(RandomForestClassifier(random_state=0), X_train, y_train, X_test, y_test)
```

Обе модели показывают близкие по значению метрики. На текущем примере  случайный лес оказался чуть лучше.

## Кластеризация клиентов

### Стандартизировать данные.

```python
# стандартизируем данные
sc = StandardScaler()
X_sc = sc.fit_transform(X)
```

### Дендрограмма

```python
# матрица расстояний функцией linkage() на стандартизованной матрице признаков
linked = linkage(X_sc, method = 'ward') 
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Иерархическая кластеризация')
plt.show()
```

Дендрограмма разделила данные на 4 кластера. Хотя в фиолетовой зоне напрашиваются 2 кластера, а не один. Поэтому скорее 5, чем 4.

### Обучить модель кластеризации на основании алгоритма K-Means. Число кластеров принять n=5.

```python
km = KMeans(n_clusters=5, random_state=0)
```

```python
# прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 4)
labels = km.fit_predict(X_sc)

# сохраняем метки кластера в поле нашего датасета
X['cluster_km'] = labels

# выводим статистику по средним значениям наших признаков по кластеру
clusters = X.groupby(['cluster_km']).mean()
print(clusters)
```

<a id='conclusion2'></a>
- 0 кластер - клиенты, которые живут не рядом с фитнес-центром, почти нет промо, небольшой абонемент - 3 мес, очень мало посещают групповые занятия;
- 1 кластер - опреленно живут рядом, есть партнерка и промо, большой абонемент - 7 мес;
- 2 кластер - определенно живут рядом, скорее нет партнерки и промо, небольшой абонемент - 3 мес, никогда не посещают групповые занятия;
- 3 кластер - определенно живут рядом, скорее нет партнерки и промо, абонемент 4.5 мес, всегда посещают групповые занятия;
- 4 кластер - в большинтсве случаев живут рядом, партнерка 50/50 промо 30/70, абонемент 4.7 мес, групповые занятия 40/60, и самое главное отличие - нет телефона!

### Построить распределения признаков для кластеров. Можно ли что-то заметить по ним

```python
X['cluster_km'].unique()
X.pivot_table(index = 'cluster_km', values = 'gender', aggfunc = 'count')
```

```python
for col in X.drop('cluster_km', axis = 1).columns:
    if len(X[col].unique()) < 4:
            report = X.reset_index().pivot_table(index = col, columns = 'cluster_km', values = 'index', aggfunc = 'count')
            report.plot(kind = 'bar', stacked = True, alpha = 0.5)
    else:
        for i in X['cluster_km'].unique():
            X.query('cluster_km == @i')[col].plot(kind = 'hist', x = col, grid=True, figsize=(7, 5), legend = True, label = i)
    plt.title(f'{col}', size = 18, pad = 25,fontdict = {'fontweight' : 'bold'})
    plt.show()
```

Вообщем-то выводы такие же, что и в [предыдущем пункте](#conclusion2).

Еще раз по основынм различиям:

- 1 кластер - опреденно имеют промо и партнерку.
- 2 кластер - нет партнерки  и промо
- 3 кластер - всегда посещают групповые занятия
- 4 кластер - нет телефона

### Для каждого полученного кластера посчитать долю оттока (методом groupby()). Отличаются ли они по доле оттока? Какие кластеры склонны к оттоку, а какие — надёжны?

```python
data['cluster_km'] = labels
data.groupby('cluster_km')['Churn'].mean()
```
