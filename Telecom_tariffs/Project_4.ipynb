## Подготовка данных

### Подготовка данных (замена типов)

В датасете `df_calls` в колонке `duration`(продолжительность разговоров) секунды должны быть округлены до минут вверх. Исправим это и изменим тип данных на `int`.

```python
df_calls['duration'] = df_calls['duration'].apply(np.ceil).astype('int')
df_calls.info()
```

Меняем тип данных с `object` на `datetime64[ns]` методом `to_datetime()`

```python
df_calls['call_date'] = pd.to_datetime(df_calls['call_date'], format = '%Y-%m-%d')
df_messages['message_date'] = pd.to_datetime(df_messages['message_date'], format = '%Y-%m-%d')
df_internet['call_date'] = pd.to_datetime(df_internet['session_date'], format = '%Y-%m-%d')
df_internet.info()
```

### Поиск аномалии в звонках, смс, и трафике.

<a id='3'></a>
### Подготовка пользовательских данных (звонки, смс, объем трафика и т.д.)

Посчитаем количество сделанных звонков и израсходованных минут разговора по месяцам

Для этого создаем столбец, где будет указан только месяц звонка.

```python
df_calls['month'] = pd.DatetimeIndex(df_calls['call_date']).month
df_calls
```

Теперь создаем сводную таблицу. Где будет видно: сколько пользователь сделал звонков, и минут использвал всего  в каждом месяце.

```python
# Создание сводной таблицы для подсчета звонков сгруппированной по user_id и month
df_calls_pt = df_calls.pivot_table(
    index = ['user_id','month'], values = 'duration', aggfunc = ['count', 'sum'])
df_calls_pt.columns = ['count_calls', 'duration']
df_calls_pt
```

Для сообщений проводим анологичные операции

```python
df_messages['month'] = pd.DatetimeIndex(df_messages['message_date']).month
```

```python
#Создание сводной таблицы дя подсчета сообщений
df_messages_pt = df_messages.pivot_table(
    index = ['user_id','month'], values = 'id', aggfunc = ['count'])
df_messages_pt.columns = ['message']
df_messages_pt.head
```

И для интернета тоже самое.

```python
df_internet['month'] = pd.DatetimeIndex(df_internet['session_date']).month
```

```python
#Создание сводной таблицы для подсчета использованного интернета по месяцам 
df_internet_pt = df_internet.pivot_table(
    index = ['user_id','month'], values = 'mb_used', aggfunc = ['sum'])
df_internet_pt.columns = ['internet']
df_internet_pt
```

<a id='4'></a>
### Создание общей таблицы для расчетов

Для создания общей таблицы используем метод `merge()`. Обьединять таблицы будем по столбцам  `'user_id', 'month'`

```python
df = df_calls_pt.merge(df_messages_pt, on = ['user_id', 'month'], how = 'outer')
df = df.merge(df_internet_pt, on = ['user_id', 'month'], how = 'outer')
df = df.reset_index().merge(df_users[['user_id', 'city', 'tariff']], on = 'user_id', how = 'outer')
df
```

В столбце `internet` округлим Мб до Гб вверх.

```python
# огругление трафика до Гб вверх
df['internet'] = (df['internet']/1024).apply(np.ceil)
```

```python
# Проверим количество пропусков появившихся после склевания таблиц
df.isna().sum()
```

В столбцах `count_calls, duration, message, internet` пропуски были потому что в данные месяцы абонент не использовал звонки, сообщения или интернет. Заменим в этих столбцах `NaN` на нули методом `fillna()`.  Есть так же 2 пропуска в столбце `month` - просто удалим строки с этим значением.

```python
# С помощью цикла пройдемся по всем столбцам сразу.
columns = ['count_calls', 'duration', 'message','internet']
for column in columns:
    df[column] = df[column].fillna(0)

#Удаление строк, где в столбце month есть пропуски
df = df.dropna(subset = ['month'])
df.info()
```

<a id='5'></a>
### Расчет помесячной выручки

Для расчета выручки создадим функцию, которая будет принимать значение строки, а на выходе будет выдавать значение выручки, которое методом `apply()` будем присваивать новому столбцу `revenue`.

```python
def get_revenue(row):
    # считаем овердрафты
    # если тариф ультра
    if row['tariff'] == 'ultra':
        
        minutes_over = (row['duration'] - 3000)*1
        message_over = (row['message'] - 1000)*1
        internet_over  = (row['internet'] - 30)*150
        tariff_price = 1950

    # если тариф смарт 
    else:
        minutes_over = (row['duration'] - 500)*3
        message_over = (row['message'] - 50)*3
        internet_over  = (row['internet'] - 15)*200
        tariff_price = 550
        
    # убираем отрицательные овердрафты
    if minutes_over < 0: minutes_over = 0
    if message_over < 0: message_over = 0
    if internet_over < 0: internet_over = 0
        
    # считаем общую выручку
    return minutes_over + message_over + internet_over + tariff_price
 
# Метод apply() применяем к датафрейму df;
# По умолчанию Pandas передаёт в функцию get_revenue() столбец.
# Чтобы на вход в функцию отправлялись строки, нужно указать параметр axis = 1 метода apply().
df['revenue'] = df.apply(get_revenue, axis = 1)
df
```

## Проверка гипотез и пояснения

Генеральные совокупности распредлены нормально.
Для проведения t-теста создадим функцию расчета p-значения и сравнения его с критическим уровнем статистической значимости.

```python
def hypothesis(x,y):
    
    # критический уровень статистической значимости
    alpha = 0.01
    
    # Проводим тест и считаем p-значение, параметру equal_var присваиваем False
    # т.к. выборки в обоих случаях разные
    results = st.ttest_ind(x, y, equal_var = False)
    print('p-значение:', results.pvalue)
    
    # Сравнениваем p-значения с заданным критическим уровнем стат. значимости
    if results.pvalue < alpha:
        return 'Отвергаем нулевую гипотезу'
    else:
        return 'Не получилось отвергнуть нулевую гипотезу'
```

### Средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются

- Нулевая гипотеза H0: Средняя выручка пользователей тарифов "Ультра" и "Смарт" равна.
- Альтернативная гипотеза H1: Средняя выручка пользователей тарифов "Ультра" и "Смарт" различается.

```python
# Расчитаем средние значения
print('cредняя выручка тарифа "Ультра": ',df_ultra['revenue'].mean())
print('cредняя выручка тарифа "Смарт": ',df_smart['revenue'].mean())

# Применяем функцию по расчету p-значения
hypothesis(df_ultra['revenue'],df_smart['revenue'])
```

p-значение стремится к нулю - это значит вероятность получить одинаковую выручку в тарифах равна 0%.

### Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

- Нулевая гипотеза H0: Средняя выручка пользователей из Москвы равна выручке пользователей из других регионов.
- Альтернативная гипотеза H1: Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

```python
# Создаем два отдельных датасета по Москве и регионам
df_moscow = df.query('city == "Москва"')
df_regions = df.query('city != "Москва"')

# Расчитаем средние значения
print('cредняя выручка в Москве: ',df_moscow['revenue'].mean())
print('cредняя выручка в регионах: ',df_regions['revenue'].mean())

# Применяем функцию по расчету p-значения
hypothesis(df_moscow['revenue'],df_regions['revenue'])# Применяем функцию по расчету p-значения
```
