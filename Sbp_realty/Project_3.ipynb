## Предобработка данных

### Заполнение пропусков

В столбцах `locality_name` и `floors_total` пропусков немного  - меньше 1%. Поэтому просто удалим строки с пропущенными значениями методом `dropna`.

```python
data = data.dropna(subset = ['locality_name']).reset_index(drop =True)
data = data.dropna(subset = ['floors_total']).reset_index(drop =True)
#data.info()
```

Проверим наличие дубликатов в данных.

```python
data.duplicated().sum()
```

Те, у кого апартаменты обычно хорошо это знают, поэтому пропуски заменим на значения `False`  с помощью метод `fillna()`

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

Если человек не указал количество балконов, скорее всего их нет. Заменим пропуски на ноль.

```python
data['balcony'] = data['balcony'].fillna(0)
```

Пропуски в `kitchen_area` и `living_area` присутсвуют скорее всего из-за того, что люди, продающие квартиры, просто не знают какие значения у этих параметров.
Заполним пропуски в `kitchen_area` и `living_area` медианным значениям по категории количества комнат.
В категории, где комнат ноль нет значений площади, поэтому посчитать медиану не получится. Вместо этого присвоим пропускам нулевые значения, т.к. скорее всего это студии и логичнее представить, что жилая и кухонная площадь равны нулю.

```python
# просмотр данных где комнат ноль
data[(data['rooms'] == 0)]
# всего 194 строчки
```

Создадим функцию для замены пустых значений для столбцов `kitchen_area`и `living_area`.

```python
def replace_area(column):
    
    for room in data['rooms'].unique():
        if room == 0:
            #в ячейках столбца где  есть пропуск и комнат ноль присваиваем площади нулевое значение.
            data.loc[(data[column].isna())&(data['rooms'] == room), column] = 0
        else:
            # Подсчет медианы для категории rooms
            median = data.loc[data['rooms'] == room, column].median()
            print(room,median)
            # Замена NaN на медиану
            data.loc[(data[column].isna())&(data['rooms'] == room), column] = median
    
    return
        
```

```python
# С помощью функции заполняем пропуски в `kitchen_area`
replace_area('kitchen_area')
```

```python
# С помощью функции заполняем пропуски в `living_area`
replace_area('living_area')
```

`cityCenters_nearest`, `airports_nearest`  - т.к. данные были заполнены автоматически на основе картографических данных, возможно пользователь не указал положение, где продается квартира.

`ceiling_height` - здесь пропусков достатчно много, надо обратить внимание разработчиков на данный факт.

В колонках `cityCenters_nearest, airports_nearest, ceiling_height` пропуски заполним медианными значениями по всем данным.

```python
# Создание списка
columns = ['cityCenters_nearest', 'airports_nearest', 'ceiling_height']
for column in columns:
    data[column] = data[column].fillna(data[column].median())
#data.info()
```

В данных по населенным пунктам, есть повторения Мурино, поселок Мурино, посёлок Мурино.
Характер таких ошибок человечкий фактор.
Уберем из названий `'поселок','посёлок', 'деревня', 'село'`.

```python
# Создание списка
names = ['поселок','посёлок', 'деревня', 'село']
# Цикл по замене слов, присутствующих в названии населенного пункта, на пустое значение
for name in names:    
    data['locality_name'] = data['locality_name'].str.replace(name, '')
```

В столбцах `parks_around3000`,`parks_nearest`,`ponds_around3000`,`ponds_nearest` тоже сть пропущенные значения, но т.к. данные этих столбцов не потребуются в исследовании - оставим их как есть.

`days_exposition`  - сколько дней было размещено объявление (от публикации до снятия).

Возможно пропуски в данном столбце связаны с тем, что обьявляения еще активны и не были сняты с публикации. Такие пропуски оставим как есть.

### Изменение типов данных

`'floors_total', 'balcony' ` имееют тип `float`, хотя логичнее должны быть целочисленные значения. Не может быть 2,5 балкона или 13,3 этажа.

В `last_price` после запятой стоят нули, и чтобы данные компактнtе смотрелись можно тоже заменить на int'

```python
columns = ['floors_total', 'balcony', 'last_price']
for column in columns:
    data[column] = data[column].astype('int')
#Проверяем
data.info()
```

## Расчёты и добавление результатов в таблицу

### Добавление цены квадратного метра

```python
# Разделим цену квартиры на общую площщадь
data['price_square'] = data['last_price']/data['total_area']
```

### Добавление дня недели, месяца и года публикации объявления 

```python
df = pd.to_datetime(data['first_day_exposition'], format = '%Y-%m-%dT%H:%M:%S')
data['weekday'] = df.dt.weekday
data['month'] = df.dt.month
data['year'] = df.dt.year
```

### Добавление этажа квартиры. Варианты — первый, последний, другой.

```python
def floor_name(row):
    floor = row['floor']
    #условия после которых функция возвращает обобщенную версию категории
    if floor == 1:
        return 'первый'
    if floor == row['floors_total']:
        return 'последний'
    return 'другой'
# Метод apply() применяем к датафрейму data;
# По умолчанию Pandas передаёт в функцию category_purpose() столбец.
# Чтобы на вход в функцию отправлялись строки, нужно указать параметр axis = 1 метода apply().
data['floor_name'] = data.apply(floor_name, axis=1)
```

### Добавление соотношения жилой и общей площади:

```python
data['living_area_ratio'] = data['living_area']/data['total_area']
```

### Добавление соотношения площади кухни к общей:

```python
data['kitchen_area_ratio'] = data['kitchen_area']/data['total_area']
data
```
